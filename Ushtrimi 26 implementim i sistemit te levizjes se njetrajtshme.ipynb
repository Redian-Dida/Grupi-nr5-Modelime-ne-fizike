```python
import matplotlib.pyplot as plt
import numpy as np
#Detyra nr.26 implementimi i sistemit te levizjes te njetrajtshme
#Punuar nga grupi 5
# Konfigurimi i parametrave fizikë
x0 = 5.0    # Pozicioni fillestar (m)
v = 2.5     # Shpejtësia konstante (m/s)

# Gjenerimi i të dhënave duke përdorur numpy për efikasitet
t = np.linspace(0, 10, 11)  # 11 pika nga 0 në 10 sekonda
x = x0 + v * t               # Ekuacioni i lëvizjes: x = x0 + v*t

# Afishimi i rezultateve në konsolë
print(f"{'Koha (s)':<10} | {'Pozicioni (m)':<15}")
print("-" * 28)
for i in range(len(t)):
    print(f"{t[i]:<10.1f} | {x[i]:<15.1f}")

# Vizualizimi i lëvizjes
plt.figure(figsize=(8, 5))
plt.plot(t, x, 'r-o', label=f'v = {v} m/s')

# Stilimi i grafikut
plt.title("Analiza e Lëvizjes Drejtvizore të Njëtrajtshme", fontsize=12)
plt.xlabel("Koha (s)")
plt.ylabel("Pozicioni (m)")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)

plt.show()
```
