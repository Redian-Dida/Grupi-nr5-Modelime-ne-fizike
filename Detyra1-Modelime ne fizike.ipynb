```python
import numpy as np
import matplotlib.pyplot as plt
import os
#Punuar nga;Redian  Dida,Klevio Shtembari,Eglant Lacka.
# Krijimi i direktorisë për figura nëse nuk ekziston
if not os.path.exists('figures'):
    os.makedirs('figures')

# --- DETYRA 3: Llogaritjet me NumPy ---

# 1. Krijimi i vektorit të kohës t në [0, 10]
t = np.linspace(0, 10, 1000)

# 2. Definimi i një funksioni determinist (Identiteti i grupit tonë)
# Përdorim një kombinim frekuencash për një pamje më "shkencore"
f_t = np.sin(2 * np.pi * 0.5 * t) * np.exp(-0.1 * t) + 0.2 * np.cos(5 * t)

# 3. Operacionet me NumPy
# Derivati numerik (df/dt)
df = np.gradient(f_t, t)

# Integrali kumulativ
dt = t[1] - t[0]
integral_kumulativ = np.cumsum(f_t) * dt

# Norma e vektorit
norma = np.linalg.norm(f_t)

# Statistikat shtesë
f_max, f_min = np.max(f_t), np.min(f_t)
f_mean = np.mean(f_t)

# Printimi i rezultateve në terminal
print(f"{'--- Rezultatet e Analizës ---':^40}")
print(f"Norma e vektorit:  {norma:.4f}")
print(f"Vlera Maksimale:   {f_max:.4f}")
print(f"Vlera Minimale:    {f_min:.4f}")
print(f"Vlera Mesatare:    {f_mean:.4f}")

# --- DETYRA 1: Figura e Stilizuar (signature_plot.png) ---

plt.figure(figsize=(10, 6), dpi=100)

# Stilizimi i vijave dhe markerëve
plt.plot(t, f_t, label='Sinjali i Grupit', color='#2E86C1', linewidth=2.5, linestyle='-')
plt.plot(t[::50], f_t[::50], 'ro', markersize=5, label='Pikat e mostrimit') # Markerët

# Konfigurimi i grid-it dhe fonteve
plt.grid(True, linestyle='--', alpha=0.6)
plt.title('Identiteti Grafik i Grupit (Analiza Shkencore)', fontsize=16, fontweight='bold', family='serif')
plt.xlabel('Koha (t)', fontsize=12, family='serif')
plt.ylabel('Amplituda f(t)', fontsize=12, family='serif')

# Paleta e ngjyrave dhe legjenda
plt.axhline(0, color='black', linewidth=1)
plt.legend(loc='upper right', frameon=True, shadow=True)

# Ruajtja e figurës
plt.savefig('figures/signature_plot.png')
print("\n[SUKSES] Figura u ruajt në: figures/signature_plot.png")
plt.show()
```
